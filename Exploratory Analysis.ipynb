```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import data
cars = pd.read_csv('completeData.csv')
cars.head()
```

First, we will take a quick look at the correlation and statistics of the numerical data

```python
cars.describe()
```

```python
_ = sns.heatmap(cars.corr(), annot=True) 
```

The strongest positive correlations to price are engineSize and year. Both mpg and mileage have a strong negative correlation.

```python
_ = sns.pairplot(cars) 
```

#### Initial observations:

Response variable is price, which is mostly grouped to the lower end.

Year: Most cars sold are from 2000 or newer. Cars from 2010 and before have almost exclusively had lower prices.

mileage: Similar to year but inverted, most cars have under a certain mileage, and those with higher mileage have lower prices.

tax: Tax seems to be certain values, possibly chosen on factors other than value of the car.

mpg: Mpg is negatively correlated to price. This is likely due to the abnormal way that mpg was calculated by whoever took this data, as mentioned in the previous notebook.

engineSize: Despite the large range of data making it seem less certain visually, larger engines are generally more expensive.

Moving on to categorical data

```python
plt.figure(figsize=(10,10))
_ = sns.violinplot(data=cars, x='make', y='price')
```

All makes are sold for mostly lower prices, but some makes top out much higher than others. These makes also have less cheaper cars.

```python
plt.figure(figsize=(10,10))
_ = sns.violinplot(data=cars, x='transmission', y='price')
```

Not having at least semi-automatic transmission seems to really limit the sale price

```python
plt.figure(figsize=(10,10))
_ = sns.violinplot(data=cars, x='fuelType', y='price')
```

Gas powered cars have the largest variation in price, likely due in part to how much more common they are. Electric cars seem to only get sold at around 20,000, which might be due to some strange sampling error. Keep in mind.

```python
plt.figure(figsize=(20,20))
_ = sns.scatterplot(data=cars, x='model', y='price', hue='make')
```
